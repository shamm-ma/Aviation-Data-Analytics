## Data-Cleaning 

```python
df_clean= df.dropna() 
df_clean.shape
```

After removing rows with missing values, the cleaned dataset contains 7,427 rows and 21 columns. Only 4 rows were removed, so the data loss is minimal.

```python
df_clean.isnull().sum()
```

All missing values have been removed from the cleaned dataset.

### Column Rename

```python
df_clean = df_clean.rename(columns={
    "carrier_ct": "carrier_delay_count",
    "weather_ct": "weather_delay_count",
    "nas_ct": "nas_delay_count",
    "security_ct": "security_delay_count",
    "late_aircraft_ct": "lateaircraft_delay_count",
    "arr_delay": "arrival_delay_min",
    "carrier_delay": "carrier_delay_min",
    "weather_delay": "weather_delay_min",
    "arr_cancelled": "cancelled_arrivals",
    "arr_diverted": "diverted_arrivals",
    "nas_delay": "nas_delay_min",
    "security_delay": "security_delay_min",
    "late_aircraft_delay": "late_aircraft_delay_min",
    "carrier": "carrier_code",
    "airport": "airport_code",
    "arr_del15": "delayed_arrivals_15m"
})
pd.set_option("display.max_columns", None)
df_clean.head()
```

## Feature Engineering
Creation of a delay rate column to measure the percentage of arriving flights that were delayed by 15 minutes or more.

```python
df_clean["delay_rate"] = df_clean["delayed_arrivals_15m"] / df_clean["arr_flights"] * 100

df_clean[[
    "carrier_name",
    "airport_name",
    "arr_flights",
    "delayed_arrivals_15m",
    "delay_rate"
]].head()
```

## On-Time Performance by Airline

```python
on_time_performance = (
    df_clean.groupby("carrier_name")
    .agg(
        total_flights=("arr_flights", "sum"),
        delayed_flights=("delayed_arrivals_15m", "sum"),
        average_delay_rate=("delay_rate", "mean"),
        total_delay_minutes=("arrival_delay_min", "sum")
    )
)

on_time_performance["on_time_flights"] = (
    on_time_performance["total_flights"] - on_time_performance["delayed_flights"]
)

on_time_performance["otp"] = (
    on_time_performance["on_time_flights"]
    / on_time_performance["total_flights"]
    * 100
)

on_time_performance_filtered = on_time_performance[
    on_time_performance["total_flights"] >= 10000
].sort_values("otp", ascending=True)

on_time_performance_filtered
```

### Airline On-time Performance Notes
- Airlines with fewer than 10,000 total flights were excluded to avoid ranking airlines with small flight volumes in order to standardize data. 
- GoJet Airlines has the lowest on-time performance, with an OTP of 71.54%.
- Frontier Airlines has the second-lowest OTP at 72.56%, followed by JetBlue Airways at 73.40%.
- Horizon Air and Mesa Airlines notably have the highest OTP rates at 83%. 

## Delay Causes 

```python
delay_causes = pd.DataFrame({
    "delay_cause": [
        "Carrier",
        "Weather",
        "NAS",
        "Security",
        "Late Aircraft"
    ],
    "delay_count": [
        df_clean["carrier_delay_count"].sum(),
        df_clean["weather_delay_count"].sum(),
        df_clean["nas_delay_count"].sum(),
        df_clean["security_delay_count"].sum(),
        df_clean["lateaircraft_delay_count"].sum()
    ]
})

delay_causes.sort_values("delay_count", ascending=False).reset_index(drop=True)
```

```python
route_delays = (
    df_clean.groupby(["airport_name", "carrier_name"])
    .agg(
        total_flights=("arrival_flights", "sum"),
        delayed_flights=("delayed_arrivals_15m", "sum"),
        total_delay_minutes=("arrival_delay_min", "sum")
    )
    .sort_values("delayed_flights", ascending=False)
    .head(5)
)

route_delays
```

## Monthly On-Time Performance

```python
monthly_performance = (
    df_clean.groupby("month")
    .agg(
        total_flights=("arr_flights", "sum"),
        delayed_flights=("delayed_arrivals_15m", "sum"),
        total_delay_minutes=("arrival_delay_min", "sum")
    )
    .reset_index()
)

monthly_performance["on_time_flights"] = (
    monthly_performance["total_flights"] - monthly_performance["delayed_flights"]
)

monthly_performance["otp"] = (
    monthly_performance["on_time_flights"]
    / monthly_performance["total_flights"]
    * 100
)

monthly_performance = monthly_performance.sort_values("otp").reset_index(drop=True)

monthly_performance
```

### Monthly On-Time Performance Notes
- July has the lowest OTP at 72.29%, making it the worst-performing month in the dataset.
- June has a very similar OTP at 72.30%, making it the second-worst month.
- September has the highest OTP at 83.85%, making it the best-performing month.
- July also has the highest number of delayed flights, with 192,888 delayed arrivals.
- The results show weaker on-time performance during June and July compared with the other months.
